Question 1: What is hypothesis testing in statistics?

Answer: Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It involves formulating assumptions (hypotheses) about a population parameter, then using statistical evidence to accept or reject these assumptions.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:

  - Null Hypothesis (H₀): A statement that there is no effect, no difference, or no relationship between variables. It represents the default or status quo assumption.

  - Alternative Hypothesis (H₁ or Ha): A statement that contradicts the null hypothesis, indicating that there is an effect, a difference, or a relationship.

Example:

  - H₀: The mean weight of apples is 150g.

  - H₁: The mean weight of apples is not 150g.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Answer: The significance level (α) is the threshold probability used to decide whether to reject the null hypothesis. Common values are 0.05, 0.01, or 0.10.

  - If the p-value ≤ α, reject H₀ (evidence supports H₁).

  - If the p-value > α, fail to reject H₀ (insufficient evidence against H₀). It represents the risk of making a Type I error (rejecting a true null hypothesis).

Question 4: What are Type I and Type II errors? Give examples of each.

Answer:

  - Type I Error (False Positive): Rejecting H₀ when it is actually true.

  Example: Concluding a new drug works when it actually doesn’t.

  - Type II Error (False Negative): Failing to reject H₀ when it is actually false.

  Example: Concluding a new drug doesn’t work when it actually does.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

Answer:

| Feature                           | Z-Test                                 | T-Test                            |
| --------------------------------- | -------------------------------------- | --------------------------------- |
| **Population Standard Deviation** | Known                                  | Unknown                           |
| **Sample Size**                   | Large (n > 30)                         | Small (n ≤ 30)                    |
| **Distribution**                  | Normal distribution                    | t-distribution                    |
| **Use Case**                      | Comparing population mean with known σ | Comparing means when σ is unknown |


Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10  # number of trials
p = 0.5  # probability of success
size = 1000  # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=np.arange(-0.5, n+1.5, 1), edgecolor='black')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python.

Answer: We will test if the mean of the sample data is equal to a hypothesized population mean (H₀: μ = 50).

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu_0 = 50  # hypothesized mean
sigma = 0.5  # assumed population standard deviation (known)
n = len(sample_data)
sample_mean = np.mean(sample_data)

# Z-test statistic
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Z-Statistic:", z_stat)
print("P-Value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject H0: There is evidence the mean is different from 50.")
else:
    print("Fail to Reject H0: No significant evidence the mean is different from 50.")
```

Question 8: Simulate data from a normal distribution and calculate the 95% confidence interval.

Answer: We will generate normally distributed data and compute the confidence interval for its mean.

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulate data
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=200)

# Calculate statistics
mean = np.mean(data)
std_err = stats.sem(data)  # standard error
conf_interval = stats.norm.interval(0.95, loc=mean, scale=std_err)

print("Mean:", mean)
print("95% Confidence Interval:", conf_interval)

# Plot
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.axvline(conf_interval[0], color='green', linestyle='--', label='Lower CI')
plt.axvline(conf_interval[1], color='green', linestyle='--', label='Upper CI')
plt.title('Normal Distribution with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

Question 9: Calculate Z-scores and visualize standardized data.

Answer: Z-score represents how many standard deviations a data point is from the mean.


```python
# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = [(x - mean) / std_dev for x in data]
    return np.array(z_scores)

# Calculate Z-scores for the same simulated data from Q8
z_scores = calculate_z_scores(data)

# Print first 5 Z-scores
print("First 5 Z-scores:", z_scores[:5])

# Plot histogram of Z-scores
plt.hist(z_scores, bins=20, color='lightcoral', edgecolor='black')
plt.title('Histogram of Z-scores')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.axvline(0, color='blue', linestyle='--', label='Mean (Z=0)')
plt.legend()
plt.show()
```
